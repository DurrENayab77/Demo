Consider the following scenario:

A sequencing machine processes 1,200 base pairs per second.

A second (newer) machine processes 1,800 base pairs per second but starts working 10 minutes later.

Both machines are sequencing the same long DNA fragment.

Q1: How long does it take the newer machine to catch up with the older one?

Q2: How many base pairs will each machine have sequenced at that point?



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40, 2500000) # start, finish, n points
```

```python
old_machine = 72000 * t                           # starts immediately at 0
```

```python

new_machine = np.where(t >= 10, 108000 * (t - 10), 0)  # starts at 10 min
```

```python
fig, ax = plt.subplots()
plt.title('Sequencing machine processes')
plt.xlabel('time (in minutes)')
plt.ylabel('base pairs sequenced')
ax.set_xlim([0, 40])
ax.set_ylim([0, 2500000])

ax.plot(t, old_machine, c='green', label='Old Machine')
ax.plot(t, new_machine, c='brown', label='New Machine')

# Mark the catch-up point
plt.axvline(x=30, color='purple', linestyle='--')  # 30 min after old started
plt.axhline(y=2160000, color='purple', linestyle='--')

plt.legend()
plt.show()
```

Two bacterial populations are being tracked:

Population A accumulates mutations at 300 per hour.

Population B accumulates mutations at 450 per hour, but it starts 2 hours later.

Write the equations for both.

When will the mutation counts be equal, and how many mutations will each have?

```python
t = np.linspace(0, 40, 3000) # start, finish, n points
```

```python
P_A = 300 * t                           # starts immediately at 0
```

```python

P_B = np.where(t >= 2, 450 * (t - 2), 0)  # starts at 10 min
```

```python
fig, ax = plt.subplots()
plt.title('Mutation rate')
plt.xlabel('time (in minutes)')
plt.ylabel('mutation count')
ax.set_xlim([0, 10])
ax.set_ylim([0, 3000])

ax.plot(t, P_A, c='green', label='Population_A')
ax.plot(t, P_B, c='brown', label='Population_B')

# Mark the catch-up point
plt.axvline(x=6, color='purple', linestyle='--')  # 30 min after old started
plt.axhline(y=1800, color='purple', linestyle='--')

plt.legend()
plt.show()
```
